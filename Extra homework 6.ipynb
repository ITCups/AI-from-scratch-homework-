```python
import numpy as np
```

```python
class LinearPredictModel:
  k = 0
  b = 0
  def predict(self, x1):
    return self.b + self.k * x1 

  def train(self,x,y,steps,L):
    length = len(x)
    for i in range(steps):
      predicts = [self.predict(stepX) for stepX in x]
      dx = (-2/length) * np.sum(x * (y - predicts))
      qx = (-2/length) * np.sum(y - predicts)
      self.k = self.k - L * dx
      self.b = self.b - L * qx
```
